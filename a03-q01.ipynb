```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt


def monte_carlo(func, x1, x2, n):
    
    x = np.arange(x1,x2,0.1)
    
    y = list()
    
    for i in range(len(x)):
        y.append(func(x[i]))
        
    max_res = max(y)
    
    xrand = x1 + np.random.uniform(x1,x2,n)
    
    yrand = 0 + np.random.random(n)*max_res
    
    
    f = np.vectorize(func)
    points_above = np.where(yrand < f(xrand))
    points_below = np.where(yrand >= f(xrand))
    
    plt.plot(x, y, color = "orange")
    
    plt.scatter(xrand[points_below], yrand[points_below], color = "black")

    plt.scatter(xrand[points_above], yrand[points_above], color = "blue")
    
    plt.legend(('points below the curve', 'points above the curve'),
           loc='lower left',
           ncol=3,
           fontsize=8)
    
    print("area of rectangle: ", max_res * (x2-x1))
    print("area under the curve: ", max_res * (x2-x1) * len(points_below[0])/n)
```

**Q: 01 (a)**

```python
def cos_func(x):
    y =  math.pow(math.cos(x), 2) + 1
    y =  math.sqrt(y)
    return y

x1 = 0
x2 = 2 
n = 10000

monte_carlo(cos_func, x1, x2, n)
```

```python
def e_func(x):
    return math.exp(x**2)

x1 = 0
x2 = 1 
n = 10000

monte_carlo(e_func, x1, x2, n)
```
