**part a**

```python
import itertools, random
import matplotlib.pyplot as plt
import numpy as np

deck = list(itertools.product(range(1,14),['Spade','Heart','Diamond','Club']))

n = 10000 

def fucn_a(n,deck):
    double_queens = 0

    for i in range(n):
        random.shuffle(deck)
        for i in range(len(deck)-1):
            if deck[i][0] == 12 and deck[i+1][0] == 12:
                double_queens += 1 
    return double_queens / n 

prob = fucn_a(n, deck)
print("prob of double queens in ",n ," no of shuffles is: ", prob)
```

```python
n = np.arange(1000, 100000, 1000)
probabilities = list()

for i in n:
    probabilities.append(fucn_a(i, deck))
```

```python
plt.figure(figsize=(10, 5))

plt.plot(n, probabilities, color = "orange")
```

**part b**

```python
n = 10000 

def func_b_a(n, deck):
    king_queen = 0
    for i in range(n):
        random.shuffle(deck) 
        
        for i in range(len(deck)-1):
           
            if deck[i][0] == 12 and deck[i+1][0] == 13:
                king_queen += 1
            
            elif deck[i][0] == 13 and deck[i+1][0] == 12:
                king_queen += 1
        
        for i in range(len(deck)-2):     
     
            if deck[i][0] == 12 and deck[i+2][0] == 13:
                king_queen += 1
            
            elif deck[i][0] == 13 and deck[i+2][0] == 12:
                king_queen += 1
            
    return king_queen / n    

probability = func_b_a(n, deck)   

print("prob of that at 1, 1 king and, 1 queen appear next to each other in shuffles: ", probability)
```

```python
n = np.arange(1000, 10000, 100)

probabilities = list() 
for i in n:
    probabilities.append(func_b_a(i, deck)) 
```

```python
plt.figure(figsize=(10, 5))
plt.plot(n, probabilities, color = "orange")
```
